```python
import pandas as pd

data = pd.read_csv('../../../../../Downloads/archive/GSAF5.csv', encoding='cp1252' )
data.head()
```

```python
data.columns
```

```python
#drop all values after row 6302 because all of them are null

data = data.drop(data.index[6302:])
```

```python
#drop all duplicate values

subset = data[['Date', 'Type', 'Country', 'Area', 'Location', 'Activity', 'Name', 'Sex ', 'Age', 'Injury', 'Fatal (Y/N)']]

subset.duplicated().any()   # devuelve True

subset[subset.duplicated()]   # obtengo los indices de los duplicados: 4688, 5709, 6295

data.drop([4688, 5709], inplace=True)
```

```python
#Create new column call case number, with unique values

data['Case Number'] = [int(e) for e in range(1, len(data)+1)]
```

```python
#Changing values from Case Number.1 to Date because from 0 to 6177 case number has the dates
data.loc[0:6177, "Date"] = data.loc[0:6177, 'CaseNumber.1']
#removing extra values from date
data['Date'] = data['Date'].str.extract('^(\d{4}\.\{2}d\.\d{2})', expand =False)
#behind the row 6177 values arent exact so 'unknown'
data.loc[6177:, 'Date'] = 'unknown'
```

```python
#eliminating 0 that is at the final of a lot of years
data['Year'] = data['Year'].astype(str).replace(to_replace=r'\.0', value = '', regex=True)
```

```python
#there are 5990 and 5991, nan values from 5992 in the columns Unname 22 and Unnamed 23 so we are going to drop them.
print(data['Unnamed: 23'].isna().sum())
print(data['Unnamed: 22'].isna().sum())

data = data.drop(['Unnamed: 23', 'Unnamed: 22'], axis = 1)
data.columns
```

```python
#Drop year since it is in data and change date to date-format
data['Date'] = pd.to_datetime(data['Date'], errors = 'coerce')
data = data.drop(['Year'], axis = 1)
```

```python
data.isnull().sum()
```

```python
bool_series = pd.isnull(data["Time"]) 
    
# filtering data 
# displaying data only with Gender = NaN 
data[bool_series] 
```

```python
#using 'unknown' in nan values of column Time
import numpy as np

data['Time'] = np.where(data['Time'].isna(), 'unknown', data['Time'])
data.isnull().sum()
```

```python
#remove spaces from columns Species, Sex

data.columns = data.columns.str.replace(' ', '')
data.columns
```

```python
#Removing unknown Age values
data['Age'] = np.where(data['Age'].isna(), 'unknown', data['Age'])
```
